```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your DataFrame (assuming it's named df)
df = pd.read_csv('mini.csv')

# Convert "Attendence %" column to strings
df['Attendence %'] = df['Attendence %'].astype(str)

# Drop rows with NaN or infinity values in the 'Roll\nNo.' column
df = df.dropna(subset=['Roll\nNo.'])
df = df[~df['Roll\nNo.'].isin([float('inf'), float('-inf')])]

# Convert 'Roll\nNo.' to integers
df['Roll\nNo.'] = df['Roll\nNo.'].astype(int)

# Define the days
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']

# Plotting
plt.figure(figsize=(10, 6))

# Iterate over each day
for day in days:
    plt.plot(df.index, df[day], label=day)

plt.xlabel('Time')
plt.ylabel('Attendance Percentage')
plt.title('Attendance Percentages for Each Day Over Time')
plt.xticks(df.index, df.index + 1)  # Set x-axis ticks to be the days
plt.legend(title='Day', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Distribution of marks obtained for each division using bar plots.
# Calculate the mean GPA for each division
division_marks = df.groupby('Division')['GPA'].mean().reset_index()
# Creating and Plotting
plt.figure(figsize=(10, 6))  # Set the figure size
plt.bar(division_marks['Division'], division_marks['GPA'], color='green')

# Set labels and title
plt.xlabel('Division')  # Set x-axis label
plt.ylabel('Mean Marks Obtained')  # Set y-axis label
plt.title('Distribution of Marks Obtained by Division')  # Set plot title

# Show grid lines on the y-axis
plt.grid(axis='y')  

# Automatically adjust subplot parameters to give specified padding
plt.tight_layout()  
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV file into a pandas DataFrame
df = pd.read_csv('mini.csv')

# Assuming the first row is the header, we can drop it
df = df.drop(0)

# Convert 'Attendence %' column to numeric
df['Attendence %'] = df['Attendence %'].str.rstrip('%').astype(float)

# Convert 'Roll No.' column to integers
df['Roll\nNo.'] = df['Roll\nNo.'].astype(int)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(df['Roll\nNo.'], df['Attendence %'], marker='o', linestyle='-')
plt.title('Attendance Percentage vs. Roll No.')
plt.xlabel('Roll No.')
plt.ylabel('Attendance Percentage')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Group by division and calculate mean marks
division_marks = df.groupby('Division')['GPA'].mean().reset_index()

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(division_marks['Division'], division_marks['GPA'], color='skyblue')
plt.title('Average Marks by Division')
plt.xlabel('Division')
plt.ylabel('Average Marks (GPA)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Extract marks column
marks = df['GPA']

# Plotting and creating a histogram
plt.figure(figsize=(10, 6))
plt.hist(marks, bins=10, color='red', edgecolor='black')
plt.title('Histogram of Marks')
plt.xlabel('Marks (GPA)')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Convert 'Attendence %' column to string
df['Attendence %'] = df['Attendence %'].astype(str)

# Plotting and creating a histogram 
plt.figure(figsize=(10, 6))
plt.hist(df['Attendence %'], bins=10, color='yellow', edgecolor='black')
plt.title('Histogram of Attendance Percentage')
plt.xlabel('Attendance Percentage')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Convert 'Attendance %' to numeric (remove '%' and convert to float)
df['Attendence %'] = df['Attendence %'].str.rstrip('%').astype(float)

# Plotting and creating a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Attendence %'], df['GPA'], color='blue', alpha=0.6)
plt.title('Scatter Plot of Attendance Percentage vs. Marks')
plt.xlabel('Attendance Percentage')
plt.ylabel('Marks (GPA)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Load CSV file into a pandas DataFrame
df = pd.read_csv('mini.csv')

# Convert 'Attendance %' to numeric (remove '%' and convert to float)
df['Attendence %'] = df['Attendence %'].str.rstrip('%').astype(float)

# Create a pivot table for heatmap and plotting
pivot_table = df.pivot_table(index='Attendence %', columns='GPA', aggfunc='size', fill_value=0)
plt.figure(figsize=(10, 8))
sns.heatmap(pivot_table, cmap='YlGnBu', annot=True, fmt='g')
plt.title('Heatmap of Attendance Percentage vs. Marks')
plt.xlabel('Marks (GPA)')
plt.ylabel('Attendance Percentage')
plt.tight_layout()
plt.show
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load CSV file into a pandas DataFrame
df = pd.read_csv('mini.csv')

# Convert 'Attendence %' column to numeric, coercing errors to NaN
df['Attendence %'] = pd.to_numeric(df['Attendence %'], errors='coerce')

# Drop rows with NaN values in the 'Attendence %' column
df.dropna(subset=['Attendence %'], inplace=True)

# Pivot the DataFrame to have 'Roll No.' as index and 'Attendence %' as columns
heatmap_data = df.pivot_table(index='Roll\nNo.', columns='Attendence %', aggfunc='mean', fill_value=0)

# Plotting
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap='YlGnBu', linewidths=0.5, linecolor='gray')
plt.title('Heatmap of Attendance Percentage vs. Roll No.')
plt.xlabel('Attendance Percentage')
plt.ylabel('Roll No.')
plt.tight_layout()
plt.show()



```

```python
# Calculate gender distribution
gender_counts = df['Gender'].value_counts()
# Plotting and creating Pie Chart
plt.figure(figsize=(8, 8))  # Set the figure size
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140, colors=['lightgreen', 'lightcoral'])
plt.title('Gender Distribution')
# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')  
# Automatically adjust subplot parameters to give specified padding
plt.tight_layout()  
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load CSV file into a pandas DataFrame
df = pd.read_csv('mini.csv')

# Replace infinite values with NaN
df.replace([np.inf, -np.inf], np.nan, inplace=True)

# Plotting with seaborn's FacetGrid
g = sns.FacetGrid(df, col='City', row='Gender', hue='Division', palette='Set1')
# Mapping histograms of 'Attendance %' to each grid element
g.map(sns.histplot, 'Attendence %')
g.set_axis_labels('Attendance Percentage', 'Count')
g.add_legend()
plt.show()
```
